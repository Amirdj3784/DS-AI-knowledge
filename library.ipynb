```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Signal Dataset.csv')
```

```python
u = np.array(df['u'])
w = np.array(df['w'])
v = np.array(df['v'])
```

```python
plt.plot(u , lw= 0.7 , color="r" , label='u c/m')
plt.plot(w , lw= 0.7 , color="b" , label='w c/m')
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(  projection='3d')
ax.plot(u , v, u)
plt.show()
```

```python
mean_v = np.mean(v)
plt.plot([],[],)
plt.plot(v , lw = 0.7 , color="black")
plt.show()
```

```python


w = np.random.rand(100)  
mean_w = np.mean(w)
k = 0.6
std_w = np.std(w)
upper_i = mean_w + k * std_w
lower_i = mean_w - k * std_w
plt.plot(w, label='w c/m', lw=0.7, color="black")
plt.plot([0, len(w)], [mean_w, mean_w], label='mean', color="red", lw=3)
plt.plot([0, len(w)], [upper_i, upper_i], lw=3, label='upper', color="blue")
plt.plot([0, len(w)], [lower_i, lower_i], lw=3, label='lower', color="green")

plt.legend()  
plt.show()
```
